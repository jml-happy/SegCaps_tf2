```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(np.squeeze(img))
plt.show()
plt.close()
plt.imshow(np.squeeze(msk), cmap='gray')
plt.show()
plt.close()
```
